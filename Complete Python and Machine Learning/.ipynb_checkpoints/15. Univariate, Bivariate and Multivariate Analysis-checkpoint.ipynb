```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
df.shape
```

## Univariate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.xlabel('Sepal Length')
plt.show()
```

## Bivariate Analysis


```python
sns.FacetGrid(df,hue="species",height=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="species",height=2)
```

## Histograms

```python
plt.hist(df['petal_length'], bins=10, range =(0,7), density=True)
```
