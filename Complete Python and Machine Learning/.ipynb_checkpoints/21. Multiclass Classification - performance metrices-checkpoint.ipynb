## Performance Metrics on Multiclass classification

```python
from sklearn import metrics
```

```python
C="Cat"
D="Dog"
F="Fox"
```

#### The precision for the cat class is the number of correctly predicted cat out of all predicted values.

#### The Recall for the cat class is the number of correctly predicted cat out of the number of actual cats.

```python
from IPython.display import Image, display

display(Image(filename="./images/my_image_4.png"))
```

```python
# True Values 
y_true = [C,C,C,C,C,C, F,F,F,F,F,F,F,F,F,F, D,D,D,D,D,D,D,D,D,D ]

# predicted values
y_pred = [C,C,F,C,D,C, F,F,F,C,F,D,F,C,F,F, D,D,C,D,D,D,F,D,D,F ]

print(metrics.confusion_matrix(y_true, y_pred)) # print the confusion matrix

print(metrics.classification_report(y_true, y_pred, digits=3 )) # print the report
```
