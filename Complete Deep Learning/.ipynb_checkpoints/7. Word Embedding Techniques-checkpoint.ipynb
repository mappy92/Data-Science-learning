## Word Embedding Techniques using Embedding Layer in Keras

```python
### Libraries USed Tensorflow> 2.0  and keras
```

```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000
```

### One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=10))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
print(model.predict(embedded_docs)[0]) # each word of the sentence converted into 10 dimension 
                                       # so each word will be represented by an array of 10 dimensions
```

```python
print(model.predict(embedded_docs))
```
